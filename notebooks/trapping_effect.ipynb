```python
import numpy as np
import matplotlib.pyplot as plt

#changable parameter
n = 1000
c0 = 0.996 #QE value at start
c1 = 1     #QE value at end
c2 = 0.7   #QE value at end

def trapping_effect(c):
    return [c0-(1-c0)*(np.log(1+1/(t[tt]+1)**c)/np.log(2)-1) for tt in range(n)]

#make time axis and QE variation
t = np.linspace(0,n/10,n)
QE1 = trapping_effect(c1)
QE2 = trapping_effect(c2)

plt.figure(figsize=(10,8))
plt.plot(t,QE1)
plt.plot(t,QE2)
plt.title('Trapping Effect (normalized)')
plt.xlabel('Time')
plt.ylabel('Quantum Efficiency')
plt.show()
```
