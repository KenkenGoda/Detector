```python
import numpy as np
import random as rd
import pandas as pd
from decimal import Decimal, ROUND_HALF_UP
import matplotlib.pyplot as plt
```

# データの取得

```python
df_spectrum = pd.read_csv('/Users/shohei/Desktop/Shohei/Fresnel/Paper/spectrum.csv')
```

# データの修正

```python
df_spectrum = df_spectrum.rename(index=df_spectrum['wavelength']).drop('wavelength', axis=1)*10**6
```

# グラフ化

```python
def Spectrum(x, y):
    [plt.plot(x, y[temp], label=temp) for temp in y.columns]
    plt.legend(loc='upper left')
    plt.xlabel('Wavelength [μm]')
    plt.ylabel('Transit depth [ppm]')
    plt.xlim(5, 20)
```

```python
plt.figure(figsize=(15,13))
plt.rcParams['font.size'] = 22
Spectrum(df_spectrum.index, df_spectrum)
plt.show()
```
