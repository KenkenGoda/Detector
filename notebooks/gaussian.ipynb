```python
import numpy as np
from scipy.stats import norm
import random as rd
import matplotlib.pyplot as plt
import time
```

```python
start = time.time()
sample1 = norm.rvs(loc=0,scale=1,size=1000000)
plt.hist(sample1,bins=100)
plt.show()
print(time.time()-start)
print(norm.fit(sample1))
```

```python
start = time.time()
sample2 = np.array([0]*1000000,float)
for kk in range(1000000):
    sample2[kk] = rd.gauss(1000000,1)
plt.hist(sample2,bins=100)
plt.show()
print(time.time()-start)
print(norm.fit(sample2))
```
