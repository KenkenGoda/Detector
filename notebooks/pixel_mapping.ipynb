```python
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns
%matplotlib inline
```

```python
temp = [2500, 3000, 3500, 4000]
data = [np.load('/Users/shohei/Desktop/Shohei/Fresnel/Detector/Paper/Simulation/Primary/pixel_map_{}.npy'.format(T)) for T in temp]
```

```python
for T in range(4):
    for i in range(1024):
        for j in range(1024):
            if data[T][i][j]<0:
                data[T][i][j] = 0.000001
```

```python
data_log = np.log10(data)
```

```python
plt.figure(figsize=(13,13))
plt.rcParams['font.size'] = 18
ax = plt.subplot(111)
plt.imshow(data[0], cmap='gray', vmin=np.min(data[0]), vmax=np.max(data[0]))
plt.xlabel('j-th pixel')
plt.ylabel('i-th pixel')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.5)
plt.colorbar(cax=cax)
```
