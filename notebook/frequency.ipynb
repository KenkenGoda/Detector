```python
import numpy as np
import pandas as pd
from scipy.fftpack import fft
import matplotlib.pyplot as plt
```

```python
gain = np.load('/Users/shohei/Desktop/Shohei/Fresnel/Paper/Transit_WOP/gain.npy')
tra = np.load('/Users/shohei/Desktop/Shohei/Fresnel/Paper/Transit_WOP/tra.npy')
sci = np.load('/Users/shohei/Desktop/Shohei/Fresnel/Paper/Transit_WOP/sci.npy')
sub = np.load('/Users/shohei/Desktop/Shohei/Fresnel/Paper/Transit_WOP/sub.npy')
data = pd.DataFrame({'gain':gain, 'tra':tra, 'sci':sci, 'sub':sub})
time = np.array([i+1 for i in range(len(data))], float)
```

```python
FT_gain = np.abs(fft(gain))
FT_tra = np.abs(fft(tra))
FT_sci = np.abs(fft(sci))
FT_sub = np.abs(fft(sub))
frequency = time[-1]/time
```

```python
data.index = time
```

```python
FT_data = pd.DataFrame({'frequency':frequency, 'FT_gain':FT_gain, 'FT_tra':FT_tra, 'FT_sci':FT_sci, 'FT_sub':FT_sub})
FT_data = FT_data[1:int(time[-1]/2)+1].rename(index=FT_data['frequency']).drop('frequency', axis=1)
```

```python
def Curve(i, data, ylabel):
    plt.subplot(i)
    plt.plot(data.index, data.values)
    plt.xlabel('Time [min]')
    plt.ylabel(ylabel)
```

```python
plt.figure(figsize=(15,4))
plt.rcParams['font.size'] = 16
Curve(131, data['tra'], 'Flux [electrons]')
Curve(132, data['gain'], 'Amplitude')
Curve(133, data['sci'], 'Flux [electrons]')
plt.subplots_adjust(wspace=0.45)
plt.show()
```

```python
Curve(data['sub'], 'Flux [electrons]')
plt.show()
```

```python
def Frequency(data, ylabel):
    plt.figure(figsize=(7, 5))
    plt.rcParams['font.size'] = 18
    plt.xscale('log')
    plt.yscale('log')
    plt.plot(data.index, data.values, 'ro', markersize=2)
    plt.xlabel('Period [min]')
    plt.ylabel(ylabel)
```

```python
Frequency(FT_data['FT_gain'], 'Amplitude')
Frequency(FT_data['FT_tra'], 'Flux [electrons]')
Frequency(FT_data['FT_sci'], 'Flux [electrons]')
Frequency(FT_data['FT_sub'], 'Flux [electrons]')
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.rcParams['font.size'] = 18
plt.xscale('log')
plt.yscale('log')
plt.plot(FT_data.index, FT_data['FT_tra'].values, 'ro', markersize=2, label='tra', c='red')
plt.plot(FT_data.index, FT_data['FT_sci'].values, 'ro', markersize=2, label='sci', c='green')
plt.plot(FT_data.index, FT_data['FT_sub'].values, 'ro', markersize=2, label='sub', c='blue')
plt.legend(loc='best')
plt.xlabel('Period [min]')
plt.ylabel('Flux [electrons]')
plt.show()
```

```python
tra_norm = tra/tra[0]
sub_norm = sub/sub[0]
```

```python
plt.figure(figsize=(15,5))
plt.rcParams['font.size'] = 18
plt.subplot(121)
plt.plot(time, tra_norm, label='model', c='green')
plt.plot(time, sub_norm, label='result', c='blue')
plt.legend(loc='right')
plt.xlabel('Time [min]')
plt.ylabel('Normalized flux')
plt.subplot(122)
plt.plot(time, tra_norm, label='model', c='green')
plt.plot(time, sub_norm, label='result', c='blue')
plt.legend(loc='lower right')
plt.xlabel('Time [min]')
plt.ylabel('Normalized flux')
plt.ylim(0.9965, 0.9968)
plt.subplots_adjust(wspace=0.35)
plt.show()
```
